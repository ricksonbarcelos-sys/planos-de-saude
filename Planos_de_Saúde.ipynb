## ***Ranqueamento de reclamações***
Listagem de todos os planos de saúde com o seu respectivo número total de reclamações. 


```
import csv

file = '/content/drive/My Drive/Colab Notebooks/dados-gerais-das-reclamacoes-por-operadora - ajustado.csv'
with open(file) as f:
  reader = csv.reader(f)
  header_row = next(reader)  
  
  planos = []
  for row in reader:
    planos.append(row[1])

# Ranqueamento de reclamações
ranqueamento = {}
for plano in planos:
    if plano in ranqueamento:
        ranqueamento[plano] += 1
    else:
        ranqueamento[plano] = 1

for item in sorted(ranqueamento, key=ranqueamento.get, reverse=True):
    print(f'{item[:50]:55}: {ranqueamento[item]}') 
```

## ***Quais planos de saúde tiveram mais reclamações***
Gráfico com os planos de saúde que tiveram acima de 1000 reclamações no total.

```
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

grupos = []
valores = []
for item in ranqueamento:
    if ranqueamento[item] >= 1000:
      grupos.append(item[:30])
      valores.append(ranqueamento[item])
    else:
      continue

fig = plt.figure(dpi=128, figsize=(8, 6))

plt.barh(grupos, valores)
plt.title("Planos com mais reclamações", fontsize=18)
plt.xlabel("número de reclamações", fontsize=16)
plt.ylabel("planos de saúde", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=8)
plt.show()
```

## ***Quantidade de beneficiários dos planos com mais reclamações***
Número de beneficiários dos planos de saúde que tiveram acima de 1000 reclamações no total.

```
import csv
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

file = '/content/drive/My Drive/Colab Notebooks/dados-gerais-das-reclamacoes-por-operadora - ajustado.csv'
with open(file) as f:
  reader = csv.reader(f)
  header_row = next(reader)  
  
  planos = []
  beneficiarios = []
  for row in reader:
    planos.append(row[1])
    beneficiarios.append(row[2])

# Quantidade de beneficiários dos planos
total_beneficiarios = {}
i = 0;
for plano in planos:
    if plano not in total_beneficiarios:
        total_beneficiarios[plano] = beneficiarios[i]
    i += 1

ranqueamento = {}
for plano in planos:
    if plano in ranqueamento:
        ranqueamento[plano] += 1
    else:
        ranqueamento[plano] = 1

for item in sorted(ranqueamento, key=ranqueamento.get, reverse=True):
    if ranqueamento[item] >= 1000:
        print(f'{item[:50]:55}: {total_beneficiarios[item]}') 

print("")

# Gráfico de beneficiários dos planos
grupos = []
valores2 = []
for item in ranqueamento:
    if ranqueamento[item] >= 1000:
      grupos.append(item[:30])
      valores2.append(ranqueamento[item])
    else:
      continue
 
fig = plt.figure(dpi=128, figsize=(8, 6))

plt.barh(grupos, valores2)
plt.title("Qtde de beneficiários por plano de saúde", fontsize=18)
plt.xlabel("número de beneficiários", fontsize=16)
plt.ylabel("planos de saúde", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=8)
plt.show()
```

## ***As datas mais recorrentes que aparecem nas reclamações***
O período contemplado nos dados refere-se à AGO/2019 a JUL/2020.

```
import csv
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

file = '/content/drive/My Drive/Colab Notebooks/dados-gerais-das-reclamacoes-por-operadora - ajustado.csv'
with open(file) as f:
  reader = csv.reader(f)
  header_row = next(reader)  
  
  datas = []
  datas_validas = ['01','02','03','04','05','06','07','08','09','10','11','12']
  for row in reader:
    if row[4][3:5] in datas_validas:
      datas.append(row[4][3:5])

# Ranqueamento de datas
ranqueamento_datas = {}
for data in datas:
    if data in ranqueamento_datas:
        ranqueamento_datas[data] += 1
    else:
        ranqueamento_datas[data] = 1

i = 0
valores3 = []
while i < 12:
  if len(valores3) > 12:
    break
  for item in ranqueamento_datas:
    if item == datas_validas[i]:
      valores3.append(ranqueamento_datas[item])
  i += 1

# Gráfico dos meses mais recorrentes
fig = plt.figure(dpi=128, figsize=(7, 5))

plt.plot(datas_validas, valores3, linewidth=5)
plt.title("Meses com mais recorrência de reclamações", fontsize=18)
plt.xlabel("Meses do ano", fontsize=16)
plt.ylabel("N° de reclamações", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=8)
plt.show()
```
